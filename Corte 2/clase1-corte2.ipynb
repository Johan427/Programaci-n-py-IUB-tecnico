```python

#2 = Escribe una función que reciba como parámetro una cadena y devuelva un diccionario con la cantidad de apariciones de cada carácter en la cadena.
def contar_caracteres(cadena):
    diccionario = {}
    for caracter in cadena:
        if caracter in diccionario:
            diccionario[caracter] += 1
        else:
            diccionario[caracter] = 1
    return diccionario

print(contar_caracteres("hola mundo"))
```

```python
#ejercicio 3
precios = {"manzana": 4000, "pera": 2000, "uva": 1000, "fresa": 3000}
while True:
   fruta = input("Ingrese el nombre de la fruta de la fruta: ").lower()
   if fruta in precios:
      cantidad = float(input("Ingrese la cantidad de la fruta vendida: "))
      total = precios[fruta] * cantidad
      print(f"El precio total de: ${total}")
   else:
       print("error: esa fruta no existe")

   otra = input("quieres hacer otra consulta? (si/no): ")
   if otra.lower() !='si':
      break
```
